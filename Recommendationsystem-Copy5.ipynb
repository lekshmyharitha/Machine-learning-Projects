```python
def pearson_correlation(x,y):
    normalizen = q.copy()
    normalizen = normalizen.fillna(0)
    #print(normalizen)
    x_profile=np.array(normalizen.loc[x])
    #print(rxs)
    y_profile=np.array(normalizen.loc[y])
    #print(rys)
    common_index=[]
    for i in range(0,normalizen.shape[1]):
        if x_profile[i]!=0 and y_profile[i]!=0:
            common_index.append(i)
    x_avg = np.sum(x_profile)/np.count_nonzero(x_profile)
    y_avg = np.sum(y_profile)/np.count_nonzero(y_profile)   
    #print(common_index)
    #print(x_avg)
    #print(y_avg)
    #numerator
    numsumval=0
    if len(common_index)>0:
        for i in common_index:
           # print("here")
            numsumval+=(x_profile[i]-x_avg)*(y_profile[i]-y_avg)
    #print(numsumval)   
    #denominator
    densumval=0
    if len(common_index)>0 and numsumval!=0:
        densumval=0
        term_1=0
        term_2=0
        for i in common_index:
            term_1 += (x_profile[i]-x_avg)**2
            term_2 += (y_profile[i]-y_avg)**2
        term1 = np.sqrt(term_1)
        term2 = np.sqrt(term_2)
        densumval = term1*term2
    #print(densumval)
    if numsumval==0:
        return 0
    else:
        return numsumval/densumval
            
```

#### Write a function NearestNbrs(U,q,k) that takes the normalized utility matrix ‘U’,the query ‘q’ as the rating vector of a user, and finds the best ‘k’ neighbours from‘U’ based on the similarity metric Pearson Correlation coefficient

```python
def NearestNbrs(U,q,k):
    similar=list()
    matrix=utility(U)
    similar_dict={}
    rating_vector=q
    num_neighbors=k
    #print(len(matrix)
    for row in matrix.index: 
         similar_dict[row]=pearson_correlation(rating_vector,row )
    sorted_similar_dict = dict(sorted(similar_dict.items(),key=operator.itemgetter(1),reverse=True))
    
    count=k
    neighbours=[]
    for k,v in sorted_similar_dict.items():
        if count!=0:
            neighbours.append((k,v))
            count-=1
    return neighbours
```
