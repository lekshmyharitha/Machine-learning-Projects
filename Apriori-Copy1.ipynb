```python
from collections import defaultdict
```

```python
def customer_items(basket,items):
        d1 = defaultdict(list)
        for i in range(0,len(basket)):
                for j in range(0,len(items)):
                        if (basket.iloc[i]['StockCode'] == items[j]):
                                
                                key=basket.iloc[i]['CustomerID']
                                value=items[j]
                                if value not in d1[key]:
                                      d1[key].append(value)
                                #d1[key].add(items[j])
                        #else:
                                #continue
        return d1
```

```python
def item(data,items):
    data_new = []
    test= customer_items(data,items)
    for i in test:
          data_new = data_new + test[i]
    set_data = set(data_new) 
#set_data.add("Customer")
    item_list = list(set_data)
    item_list = sorted(item_list)
    return item_list
```

```python
def data_find(data,item_list):
    test= customer_items(data,item_list)
    #print(test)
    df_Total_data = []
    for i in test:
            df_data = []
            df_data.append(i)
            for j in range(len(item_list)):
                if item_list[j] in test[i]:
                     df_data.append(1)
                else:
                    df_data.append(0)  
            df_Total_data.append(df_data)
    item_sort=item(data,item_list)
    column_names = ["Customer"] + item_sort
    matrix = pd.DataFrame(df_Total_data,columns=column_names)
    return matrix
```

```python
#first phase
def frequent_set(basket,items,minimum_support ):
    matrix=data_find(basket,items)
    matrix_new=matrix.copy()
    matrix_new=matrix_new.drop('Customer',axis='columns')
    q=(matrix_new!= 0).sum()
    mask=(matrix != 0).sum()>= minimum_support 
    fset=q[mask]
    #fset_new = pd.DataFrame(fset,index=None)
    fset_new=fset.to_frame()
    #fset_new2.index = range(len(fset_new))
    #fset_new.set_index(['Item', 'Support'])
    first=fset_new.index.values
    p=first.tolist()
    fset_new2=pd.DataFrame({'Item':p,'Support':fset_new.loc[:,0]})
    fset_new2.reset_index(drop=True, inplace=True)
    trans_df = fset_new2.set_index("Item").T
    freq_dict=trans_df.to_dict('list')
    return fset_new2,freq_dict
```

# Frequent k-itemset.- L(k)  with apriori
a. Given a support threshold ‘s’, write a function to find the k-itemset having
support greater than ‘s’
b.Candidate itemset
Given two frequent k-itemset, L(k), generate L(k+1)

```python
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

```

```python
def frequent_item_set(basket,items):
    matrix=data_find(basket,items)
    matrix_new=matrix.copy()
    matrix_new=matrix_new.drop('Customer',axis='columns')
    #print(matrix_new)
    basket_sets = matrix_new.applymap(encode_units)
    frq_items = apriori(basket_sets, min_support = 0.05, use_colnames = True) 
    return frq_items
```

```python
frq_items=frequent_item_set(basket_France,items)
frq_items
```

Given an itemset with cardinality ‘T’, and confidence threshold ‘c’, write a function
to output the possible association rules with confidence greater than ‘c’.


```python
def association_rules(T,C,items,basket):
       
        required_items= [x for index, x in enumerate(items) if index < T]
        frq_items=frequent_item_set(basket,required_items)
        from mlxtend.frequent_patterns import apriori, association_rules 
        rules = association_rules(frq_items, metric="lift",min_threshold = C)
        return rules
```

```python
items=basket_France['StockCode']
items=items.values.tolist()
rules=association_rules(5,1,items,basket_France)
rules
```
